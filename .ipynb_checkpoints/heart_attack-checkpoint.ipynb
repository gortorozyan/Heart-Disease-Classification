```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from Handling_Outliers import Handling_Outliers
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv('Heart Attack.csv')
```

```python
df
```

```python
laben = LabelEncoder()
```

```python
df['class'] = laben.fit_transform(df['class'])
```

```python
cor_y = df.corr()['class']
```

```python
stay_X = cor_y[cor_y >= 0.2].index[:-1]
remove_X = cor_y[cor_y < 0.2].index
```

```python
for i in remove_X:
    df.drop(f'{i}', axis=1, inplace=True)
```

```python
X = df.drop('class', axis=1)
y = df['class']
```

```python
for i in stay_X:
    Handling_Outliers(X, stay_X)
```

```python
stand = StandardScaler()
```

```python
X = stand.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
```
